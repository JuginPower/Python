```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('SALES')

# plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1,2,2) # First parameter are the wished rows, second are the columns
plt.plot(x,y)
plt.title('INCOME')

plt.suptitle('My Shop')
plt.show()
```

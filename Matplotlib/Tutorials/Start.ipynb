```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([1, 3, 5, 8])
ypoints = np.array([3, -1, 6, 2])

plt.plot(xpoints, ypoints, label='x-y points')
plt.legend()
plt.show()
```

```python
y1 = np.array([3,1,4,2,9,5])
y2 = np.array([8,5,7,4,3,0])

plt.plot(y1, marker='D', label='y1', c='k')
plt.plot(y2, marker='*', label='y2', c='r')

plt.legend()
plt.show()
```

```python
points = np.array([3,1,4,2,9,5])
plt.plot(points, '*--g', label='points', ms=20, mec='#93d54c', mfc='hotpink') # Marker, line and color fmt-parameters('*--m').
plt.legend()
plt.show()
```

```python
points = np.array([3, 8, 1, 10])

plt.plot(points, ls ='--', c='r', lw=20.5)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

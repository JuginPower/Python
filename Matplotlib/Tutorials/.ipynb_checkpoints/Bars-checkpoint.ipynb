```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.subplot(1,2,1)
plt.barh(x, y, height=0.5)

x = np.array(['Apples', 'Bananas'])
y = np.array([400, 250])
plt.subplot(1,2,2)
plt.bar(x, y)

plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color="r", width=0.5)
plt.show()
```

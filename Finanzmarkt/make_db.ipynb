```python
import pandas as pd
import sqlite3
```

```python
from sqlalchemy import inspect, create_engine
engine = create_engine('sqlite:///Indizes.db')
inspector = inspect(engine)
print(inspector.get_table_names())
```

```python
conn = sqlite3.connect('Indizes.db')
curs = conn.cursor()
conn.commit()
```

```python
df = pd.read_csv('data/dax.csv', encoding='utf-8')
df.dropna(inplace=True)
df.drop(columns=['Adj Close'], inplace=True)
print(df.to_string())
print(df.info())
```

```python
df.to_sql('dax', conn, if_exists='append', index=False)
```

```python
curs.execute("CREATE TABLE pointandfigure (Date text, Reversale text, Point int, Scale real, Symbol text)")
conn.commit()
```

```python
curs.execute("DROP TABLE pointandfigure")
conn.commit()
```

```python
sql = "SELECT * FROM dax_pf"
df = pd.read_sql(sql, conn)
print(df.info())
```
